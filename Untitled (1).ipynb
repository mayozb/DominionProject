```python
import pandas as pd
import os
from haversine import haversine
```

```python
df=pd.read_csv("LatLong.csv", header =0)
```

```python
df2=df[:42]
```

```python
df2
```

```python
it = df2.iterrows() 
```

```python
locations = {}
for l in it:
    locations[l[1][0]]=(l[1][1].LAT, l[1][2].LONG )
```

```python
mat =[]
for df2_0 in locations:
    row =[]
    for df2_1 in locations:
        distance = haversine(df2_0, df2_1, unit='mi')
        row.append(distance)
    mat.append(row)
mat
```
