```python
import pandas as pd
import os
from haversine import haversine
from math import radians, cos, sin, asin, sqrt
```

```python
data=pd.read_csv("LatLong.csv")
#print(data.head)
```

```python
for i in data.index:
    Lat = data.loc[i,'LAT']
    Long = data.loc[i,'LONG']
    PlantID = data.loc[i,'PLANT']
    haversinesource = (Lat, Long)
    for j in data.index:
        LatTarget = data.loc[j,'LAT']
        LongTarget = data.loc[j,'LONG']
        PlantIDTarget = data.loc[j,'PLANT']
        haversinetarget = (LatTarget, LongTarget)
        if PlantID !=PlantIDTarget:
            x = haversine (haversinesource, haversinetarget, unit='mi')
            print(haversinesource, "_", haversinetarget, "_", x)
            data.at[j,PlantID] = x
        data.to_csv('DominionDistance.csv')
```
